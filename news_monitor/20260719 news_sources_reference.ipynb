## 6. GDELT Project — Global Database of Events, Language, and Tone

The **[GDELT Project](https://www.gdeltproject.org/)** is the largest open-access database of global news and events — monitoring broadcast, print, and web news worldwide in **65 languages**, updated every 15 minutes. It's the ideal complement to the text-only news sources above for large-scale social monitoring.

> **ℹ️ Spiderfood:** Many GDELT pages (especially `data.html#rawdatafiles` and the `events/` index) are designed primarily as **"spider food"** — raw index pages optimized for web crawlers and automated scripts, not for human reading. They list thousands of raw file links that bots ingest. Use HTTP HEAD requests to probe for the latest files rather than trying to browse these pages manually.

| # | Resource | URL | Type | Rep | Notes |
|---|----------|-----|:---:|:---:|-------|
| 1 | **GDELT Raw Data Files** 🕷️ | [gdeltproject.org/data.html](https://www.gdeltproject.org/data.html#rawdatafiles) | Spiderfood Index | ⭐9 | Master index of all GDELT datasets — **designed for crawlers**, not humans |
| 2 | **GDELT Events Index** 🕷️ | [data.gdeltproject.org/events/](http://data.gdeltproject.org/events/index.html) | Spiderfood Index | ⭐9 | Plain directory listing of 15-min CSV files — **spider food** |
| 3 | **GDELT 2.0 Blog** | [blog.gdeltproject.org/...](https://blog.gdeltproject.org/gdelt-2-0-our-global-world-in-realtime/) | Docs | ⭐9 | Human-readable introduction to GDELT 2.0 real-time event extraction |

### 📡 Key GDELT Data Files (Direct Download)

| File Pattern | Description | Update Frequency |
|---|---|---|
| `YYYYMMDD.export.CSV.zip` | **Events** — structured CAMEO-coded event data (Actor1, Actor2, EventType, Goldstein scale) | Every 15 min |
| `YYYYMMDD.gkg.csv.zip` | **GKG (Global Knowledge Graph)** — entities, themes, locations, sentiment from each article | Every 15 min |
| `YYYYMMDD.mentions.CSV.zip` | **Mentions** — links events to source article URLs | Every 15 min |
| `MASTERURLS.TXT` | Recent article URLs used by GDELT | Daily |

### 🔗 GDELT Endpoint Pattern

```
http://data.gdeltproject.org/gdeltv2/YYYYMMDDHHMMSS.export.CSV.zip
```
Replace `YYYYMMDDHHMMSS` with the timestamp. New files appear every 15 minutes.

### 📊 Why GDELT for Social Monitoring?

- **Scale:** Processes hundreds of thousands of articles daily from global news sources
- **Structure:** Pre-parsed with CAMEO event codes, actors, geolocation, and Goldstein sentiment scores
- **Free & Open:** No API key, no rate limits, direct HTTP download
- **Complementary:** Combine with the text-only sources above for breadth (GDELT) + depth (text-only sites)
- **Language coverage:** Monitors news in 65 languages with translation

````python
# Cell 3: Define Fetch & Parse Functions

HEADERS = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/120.0.0.0 Safari/537.36"
    )
}

TIMEOUT = 15  # seconds


def fetch_and_parse(url, name):
    """
    Fetch a text-only news URL and extract headline-like content.
    Returns a dict with status, headlines, and metadata.
    """
    result = {
        "source_name": name,
        "url": url,
        "timestamp": datetime.datetime.now().isoformat(),
        "status": "unknown",
        "headlines": [],
        "error": None,
    }
    
    try:
        resp = requests.get(url, headers=HEADERS, timeout=TIMEOUT)
        resp.raise_for_status()
        result["status"] = resp.status_code
        result["content_length"] = len(resp.text)
        
        soup = BeautifulSoup(resp.text, "html.parser")
        
        # Remove script/style tags
        for tag in soup(["script", "style", "nav", "footer", "header"]):
            tag.decompose()
        
        # Strategy 1: Extract from common headline tags (h1, h2, h3, a)
        headlines = []
        
        # Look for <a> tags with article-like content
        for tag in soup.find_all(["h1", "h2", "h3", "h4"]):
            text = tag.get_text(strip=True)
            if text and len(text) > 10 and len(text) < 300:
                headlines.append(text)
        
        # Also collect links with substantial text
        for tag in soup.find_all("a"):
            text = tag.get_text(strip=True)
            href = tag.get("href", "")
            if text and len(text) > 20 and len(text) < 200 and not href.startswith("#"):
                headlines.append(text)
        
        # Deduplicate while preserving order
        seen = set()
        unique_headlines = []
        for h in headlines:
            if h not in seen:
                seen.add(h)
                unique_headlines.append(h)
        
        result["headlines"] = unique_headlines[:30]  # limit to 30 headlines
        
    except requests.exceptions.Timeout:
        result["status"] = "timeout"
        result["error"] = f"Request timed out after {TIMEOUT}s"
    except requests.exceptions.ConnectionError:
        result["status"] = "connection_error"
        result["error"] = "Could not connect to server"
    except requests.exceptions.HTTPError as e:
        result["status"] = f"http_error"
        result["error"] = str(e)
    except Exception as e:
        result["status"] = "parse_error"
        result["error"] = str(e)
    
    return result


print("✓ fetch_and_parse() function defined")
print("  - User-Agent header configured")
print(f"  - Timeout: {TIMEOUT}s")
print("  - Returns: dict with source_name, url, timestamp, status, headlines, error")
````

````python
# Cell 3b: GDELT — Fetch Today's Global Event Data
# Downloads the latest GDELT 2.0 Events file and extracts top headlines/counts.
# GDELT files are named by 15-min timestamps, e.g. 20260719150000.export.CSV.zip

def fetch_gdelt_today():
    """
    Fetch the most recent GDELT 2.0 Events data for today.
    Returns a DataFrame with structured event data and a summary dict.
    """
    today = datetime.date.today()
    date_str = today.strftime("%Y%m%d")
    
    # GDELT Events data URL pattern
    gdelt_base = "http://data.gdeltproject.org/gdeltv2"
    
    # Try the latest 15-min file (multiple attempts back through today)
    results = []
    now = datetime.datetime.utcnow()
    
    # Try the last few time slots
    for attempt in range(8):  # Try up to 2 hours back
        test_time = now - datetime.timedelta(minutes=15 * attempt)
        time_str = test_time.strftime("%Y%m%d%H%M%S")
        url = f"{gdelt_base}/{time_str}.export.CSV.zip"
        
        try:
            resp = requests.head(url, timeout=10)
            if resp.status_code == 200:
                latest_url = url
                latest_time = time_str
                break
        except:
            continue
    else:
        # Fallback: try daily aggregate
        latest_url = f"{gdelt_base}/{date_str}.export.CSV.zip"
        latest_time = f"{date_str}000000"
        try:
            resp = requests.head(latest_url, timeout=10)
            if resp.status_code != 200:
                return None, {"error": f"No GDELT files found for {date_str}"}
        except:
            return None, {"error": "Could not reach GDELT servers"}
    
    # Download the zip
    try:
        print(f"📡 Downloading GDELT Events: {latest_url}")
        resp = requests.get(latest_url, timeout=60)
        resp.raise_for_status()
        
        # Read CSV from zip
        with zipfile.ZipFile(io.BytesIO(resp.content)) as z:
            csv_name = z.namelist()[0]
            with z.open(csv_name) as f:
                df = pd.read_csv(f, sep='\t', dtype=str, on_bad_lines='skip', nrows=50000)
        
        # GDELT 2.0 Columns (key ones):
        cols = df.columns.tolist()
        
        # Extract useful columns if they exist
        event_count = len(df)
        
        # Get top event types
        if 'EventBaseCode' in df.columns:
            top_events = df['EventBaseCode'].value_counts().head(10).to_dict()
        else:
            # GDELT 2.0 uses numeric columns
            event_col = cols[26] if len(cols) > 26 else None  # EventBaseCode column
            if event_col:
                top_events = df[event_col].value_counts().head(10).to_dict()
            else:
                top_events = {}
        
        # Get top countries mentioned
        actor1_country_col = cols[7] if len(cols) > 7 else None  # Actor1CountryCode
        actor2_country_col = cols[17] if len(cols) > 17 else None  # Actor2CountryCode
        
        country_counts = {}
        if actor1_country_col:
            country_counts = df[actor1_country_col].value_counts().head(10).to_dict()
        
        # Goldstein Scale (sentiment) — column 30 in GDELT 2.0
        goldstein_col = cols[30] if len(cols) > 30 else None
        avg_sentiment = None
        if goldstein_col:
            try:
                avg_sentiment = df[goldstein_col].dropna().astype(float).mean()
            except:
                pass
        
        summary = {
            "file": latest_time,
            "url": latest_url,
            "total_events": event_count,
            "top_event_types": top_events,
            "top_countries": country_counts,
            "avg_goldstein_score": round(avg_sentiment, 3) if avg_sentiment is not None else None,
            "timestamp": datetime.datetime.now().isoformat()
        }
        
        return df, summary
    
    except Exception as e:
        return None, {"error": str(e), "url": latest_url}

# ---- Run GDELT Fetch ----
print("=" * 60)
print("🌍 GDELT 2.0 — Global Event Data Fetch")
print("=" * 60)
print()

gdelt_df, gdelt_summary = fetch_gdelt_today()

if gdelt_df is not None:
    print(f"\n✅ GDELT Data Loaded Successfully!")
    print(f"   File: {gdelt_summary['file']}")
    print(f"   Total events in batch: {gdelt_summary['total_events']:,}")
    print(f"   Avg Goldstein Score: {gdelt_summary['avg_goldstein_score']} (negative = conflict, positive = cooperation)")
    
    if gdelt_summary['top_event_types']:
        print("\n📊 Top Event Types (CAMEO Codes):")
        print("   025 = Appeal for cooperation | 046 = Engage in negotiation")
        print("   190 = Use conventional military force | 120 = Reject")
        print("   010 = Make statement | 020 = Make an appeal")
        for code, count in list(gdelt_summary['top_event_types'].items())[:5]:
            print(f"   Code {code}: {count:,} events")
    
    if gdelt_summary['top_countries']:
        print("\n🌎 Top Countries (Actor1):")
        for country, count in list(gdelt_summary['top_countries'].items())[:5]:
            print(f"   {country}: {count:,} events")
else:
    print(f"\n⚠️ Could not fetch GDELT data: {gdelt_summary.get('error', 'Unknown error')}")

print()
print("🔗 GDELT Resources:")
print("   https://www.gdeltproject.org/data.html#rawdatafiles")
print("   http://data.gdeltproject.org/events/index.html")
print("   https://blog.gdeltproject.org/gdelt-2-0-our-global-world-in-realtime/")
````

````python
# Cell 4: Fetch and Parse All Sources
# Loops through the source list with a polite delay between requests.
# Skips known-broken & non-text-only sources for efficiency.

DELAY_BETWEEN_REQUESTS = 2  # seconds
SKIP_BROKEN = True  # Set to False to attempt all URLs

results = []
total = len(SOURCES)

print(f"🚀 Starting to fetch {total} sources...\n")

for i, (name, url, st, cat, text_only, verified, reputation) in enumerate(SOURCES, 1):
    # Optionally skip known-broken sources
    if SKIP_BROKEN and reputation == 0:
        print(f"[{i}/{total}] ⏭ SKIPPING: {name} ({url}) — known broken/inactive")
        result = {
            "source_name": name, "url": url, "timestamp": None,
            "status": "skipped_broken", "headlines": [], "error": "Known broken/inactive"
        }
        result["category"] = cat
        result["source_type"] = st
        result["text_only"] = text_only
        result["verified"] = verified
        result["reputation"] = reputation
        results.append(result)
        continue
    
    print(f"[{i}/{total}] Fetching: {name} ({url}) [{st}]")
    result = fetch_and_parse(url, name)
    result["category"] = cat
    result["source_type"] = st
    result["text_only"] = text_only
    result["verified"] = verified
    result["reputation"] = reputation
    results.append(result)
    
    headline_count = len(result["headlines"])
    if result["error"]:
        print(f"  ⚠ Error: {result['error']}")
    else:
        print(f"  ✓ Retrieved {headline_count} headline(s) ({result.get('content_length', 'N/A')} bytes)")
    
    # Polite delay between requests
    if i < total:
        time.sleep(DELAY_BETWEEN_REQUESTS)

print(f"\n✅ Finished fetching all {total} sources.")
````

````python
# Cell 5: Store and Summarize Results

# Build DataFrame
df = pd.DataFrame(results)

# Summary columns
summary = df[["source_name", "category", "source_type", "text_only", "verified", "reputation", "status", "headlines"]].copy()
summary["headline_count"] = summary["headlines"].apply(len)
summary["first_headline"] = summary["headlines"].apply(
    lambda x: x[0] if len(x) > 0 else "(no headlines)"
)

print("📊 SUMMARY: Text-Only News Sources Scan")
print(f"   Scan timestamp: {datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"   Total sources attempted: {len(summary)}")
print(f"   Successful (HTTP 200):   {len(summary[summary['status'] == 200])}")
print(f"   True text-only sources:  {len(summary[summary['text_only'] == True])}")
print(f"   Pre-verified working:    {len(summary[summary['verified'] == True])}")
print(f"   Avg reputation score:    {summary['reputation'].mean():.1f}/10")
print()

# Display the summary table sorted by reputation (highest first)
display_df = summary.sort_values("reputation", ascending=False)
display_df = display_df[["source_name", "category", "reputation", "text_only", "status", "headline_count", "first_headline"]]
display_df.columns = ["Source", "Category", "⭐Rep", "Text-Only?", "HTTP", "# Headlines", "First Headline"]
display_df.index = range(1, len(display_df) + 1)

print(display_df.to_string())

# Optional: Export to CSV
csv_filename = f"news_scan_{datetime.date.today().isoformat()}.csv"
df.to_csv(csv_filename, index=False)
print(f"\n💾 Full results exported to: {csv_filename}")

# Show reputation distribution
print("\n📊 Reputation Score Distribution:")
for score in range(10, -1, -1):
    count = len(summary[summary['reputation'] == score])
    if count > 0:
        bar = "█" * count
        label = ""
        if score >= 8:
            label = "Excellent"
        elif score >= 6:
            label = "Good"
        elif score >= 4:
            label = "Fair"
        elif score > 0:
            label = "Caution"
        else:
            label = "Broken"
        print(f"  ⭐{score}/10 [{label:10s}]: {bar} ({count})")

# Highlight top-rated text-only sources
print("\n🏆 Top-Rated Verified Text-Only Sources (⭐8+):")
top = summary[(summary["text_only"] == True) & (summary["reputation"] >= 8)]
for _, row in top.iterrows():
    print(f"  ⭐{row['reputation']}/10 — {row['source_name']}")
````
